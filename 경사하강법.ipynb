```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [[2,81], [4, 93], [6, 91], [8,97]]

x = [i[0] for i in data]
y = [i[1] for i in data]

plt.figure(figsize = (8,5))
plt.scatter(x,y)
plt.show()
```

```python
x_data = np.array(x)
y_data = np.array(y)

w = 0
b = 0

lr = 0.02
epochs = 3001

#경사하강법 시작
for i in range(epochs):
    y_pred = w * x_data + b
    error = y_data - y_pred

    # 오차 함수를 a로 미분한 값
    a_diff = -(2/len(x_data))*sum(x_data*(error))
    b_diff = -(2/len(x_data)) * sum(y_data - y_pred)

    w = w - lr * a_diff
    b = b- lr * b_diff

    if i % 100 == 0:
        print("epoch = %.f, w = %.04f , b = %.04f" %(i,w,b))

```

```python
y_pred = w * x_data + b
plt.scatter(x,y)
plt.plot([min(x_data),max(x_data)], [min(y_pred),max(y_pred)])
plt.show()
```
